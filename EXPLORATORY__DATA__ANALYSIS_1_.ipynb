```python
#box plot to find outliers
for column in df.columns:
  plt.figure(figsize=(5, 5))
  sns.boxplot(x=df[column])
  plt.title(f'Box Plot for {column}')
  plt.show()
```

```python
#filling null values with median as it hvae many outliers
for col in df.columns:
    df=df.fillna(df[col].median())
```

```python
#checking for duplicates
df.duplicated().sum()
```

```python
#identifing the duplicates
df[df.duplicated()]
```

```python
#dropping the duplicate values
df=df.drop_duplicates()
```

```python
#standardization the data
from sklearn.preprocessing import StandardScaler

# Fit and transform the numerical columns
scaler = StandardScaler()
df[df.columns] = scaler.fit_transform(df[df.columns])
```

```python
#Identifing the outliers using quartile
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Identifying outliers
outliers = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)

# Printing the row numbers of outliers
print("Outlier Row Numbers (IQR Method):")
print(df.index[outliers].tolist())
```

```python
#Dropping the outliers
df=df.drop(df.index[outliers])
```

# 2.	Statistical Summary:

```python
#Statistical Summary of all columns
for column in df.columns:
    print(f"Statistical Summary for {column}:")
    print(f'mean: {df[column].mean()}')
    print(f'meadian: {df[column].median()}')
    print(f'standard deviation: {df[column].std()}')
    print(f'skewness: {df[column].skew()}')
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    print(f"IQR : {IQR}, Q1 : {Q1} , Q3 : {IQR}", )
    print()
```

# 3.	Data Visualization:

```python
# Histogram of to visualize the distributions of various numerical variables.
for column in df.columns:
    fig=plt.figure(figsize=(5, 5))
    sns.histplot(df[column], kde=True )
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```
